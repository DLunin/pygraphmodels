```python
from copy import copy, deepcopy
from functools import reduce
import random
def tile_network(dgm, n_tiles=2, max_pa=2):
    def make_tile(dgm, tile_id):
        def new_name(node):
            return '{}_{}'.format(str(node), tile_id)
        result = deepcopy(dgm)
        for node in result.nodes():
            result.rename_variable(node, new_name(node))
        return result
    tiles = [make_tile(dgm, i) for i in range(1, n_tiles+1)]
    topsorted = reduce(list.__add__, map(nx.topological_sort, tiles))
    sources = [[node for node in tile.nodes() if tile.in_degree(node) == 0] for tile in tiles]
    selected = [np.random.choice(tile_sources) for tile_sources in sources]
    result = reduce(gm.DGM.compose, tiles)
    for node in selected:
        idx = topsorted.index(node)
        pa = random.sample(topsorted[:idx], min(idx, max_pa))
        for parent in pa:
            result.add_edge(parent, node)
    
    for node in selected[1:]:
        print(node)
        
    return result, tiles
```
